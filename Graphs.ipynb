# Load the data analysis libraries


```python
import numpy as np #for numbers
import matplotlib.pyplot as plt #for graphs
```

Get the data/generate data

```python
rainfall=[67,20,10,10,10,26]
months=['jan','feb','mar','apr','may','jun']
```

Visualize the data

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
plt.bar(months,rainfall,color='#1f77b4')
```

```python
# f=minimum 6 characters
# 0.1 ans upto 1 decimal point
plt.pie(
    rainfall,
    labels=months,
    autopct='%.1f%%',
)
plt.show()
```

```python
plt.pie(
    rainfall,
    labels=months,
    autopct='%.1f%%',
    wedgeprops={'width':0.4},
    pctdistance=.8
)
plt.show()
```

```python
plt.pie(
    rainfall,
    labels=months,
    autopct='%.1f%%',
    wedgeprops={'width':0.4, 'edgecolor': '#6666ff', 'linewidth':2},
    pctdistance=.8
)
plt.show()
```

```python
plt.boxplot(rainfall,
            vert=False,
            showmeans=True,
            meanline=True,
            meanprops={'color':'red'}
            )            
plt.show()
```
